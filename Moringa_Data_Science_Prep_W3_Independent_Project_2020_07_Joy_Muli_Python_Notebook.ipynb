DATA CLEANING, SELECTING & SORTING


```
#The Telcom_dataset table has some misspelled columns 
#Renaming the column
df2.rename(columns={'PRODUTC' : 'PRODUCT', 'DATETIME' : 'DATE_TIME'}, inplace=True)
df2.head()
```

```
#Telcome_dataset2 has some mispspelled columns
df3.rename(columns={'DW_A_NUMBER': 'DW_A_NUMBER_INT', 'DW_B_NUMBER': 'DW_B_NUMBER_INT'}, inplace=True)
df3.head()
```

```
#Telcom_dataset3 also has some misspelled columns
df4.rename(columns={'CELLID': 'CELL_ID', 'SIET_ID': 'SITE_ID'}, inplace=True)
df4.head()
```

```
#Telcome_dataset, Telcom_dataset2 and Telcom_dataset3 all have the same columns
#Concating the columns
frames = [df2, df3, df4]
df_all = pd.concat(frames)
df_all.head()
```

```
#Checking and dropping duplicates
df_all.drop_duplicates()
df_all
```

```
# Map the lowering function to all column names
# Uncomment the line below after running previous lines
df_all.columns = map(str.lower, df_all.columns)
df_all
```

```
#Dropping unnecessary columns
df_all = df_all.drop(columns=['cell_on_site', 'dw_a_number_int', 'dw_b_number_int', 'country_a', 'country_b'])
df_all
```

```
#Rearranging the column names on cells_geo
df5 = pd.read_csv('cells_geo.csv', delimiter=';')
df5.head()
```

```
df5.rename(columns={'SITE_CODE': 'SITE_ID'}, inplace=True)
df5.head()
```

```
# Map the lowering function to all column names
# Uncomment the line below after running previous lines
df5.columns = map(str.lower, df5.columns)
df5
```

```
#Dropping unnecessary columns 
df5 = df5.drop(columns=['localisation', 'decoupzone', 'zonename', 'longitude', 'latitude', 'region', 'area'])
df5
```

```
#Merging df5 and df_all
df_study = pd.merge(df_all, df5, on='site_id')
df_study
```

```
#Most used cities for 3 days
df_study['villes'].value_counts()
```

```
#Sorting the rows
df_study = df_study.sort_values(by='value', ascending=0)
df_study
```

```
# Create a groupby variable that groups value by product
groupby_product = df_study['value'].groupby(df_study['product'])
groupby_product
# Descriptive statistics by group
df_study['value'].groupby(df_study['product']).describe()
```
